    1.Matrix Addition
  
Rule :Two Matrices can be added only if the have the same shape,that,is both the matrix should have the same number of rows and columns.

Rule :Two Matrices can be subtracted only if the have the same shape,that,is both the matrix should have the same number of rows and columns.

    3.Multiplying a Matrix by a Scaler

Rule : The number of columns in the first matrix should be equal to the number of rows in the second matrix.

   The resultant matrix will have the same number of rows as the first matrix & the same number of columns as the second matrix

# ML Use Case 4: Heart Disease Prediction :-

              Work Flow

Heart Data ---->Data pre processing --->Train Test split ---->Logistic Regression model

    New data ---->Trained Logistic Regression model ---> Healthy (or) Heart Defect

 

Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing 

```python
heart_data=pd.read_csv('/content/heart_disease_data.csv')
```

```python
heart_data
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# print last 5 rows of the dataset
heart_data.tail()
```

```python
# number of rows and columns in the dataset
heart_data.shape
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking the missing value
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of target variables
heart_data['target'].value_counts()
```

1 ---> Defective Heart
0 ---> Healthy Heart

splitting the Features and Target

```python
X=heart_data.drop(columns='target', axis=1)
Y=heart_data['target']

# whenever dropping the column then you have to mention axis=1  and if you drop row from dataset then mention axis=0
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
#In simple word, it ensures that the splitted data have at least some similarity between the train and test data. Suppose a scenario 
    #in which you have 100 binary(0,1) datasets of a cancer patient and you have splitted it into 80% as train_data and 20% as test_data.
    #And by default in test_data, you get 15 values as " 1 " and 5 values as 0, So it will be a quite biased data_set. In order to avoid 
    #this situation, we use "stratify" to provide a set of common relationship between training and testing dataset.(example: 11 values 
    #as "1" and 9 values a "0")
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training :

here target value is in binary 0 and 1 from,so here we will Logistic Regression because this model is best for binary.

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the Logistic Regression model with Training data
model.fit(X_train,Y_train)
```

Model Evaluation 

Accuracy Score

```python
# accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("accuracy on training data",training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("accuracy on test data",test_data_accuarcy)
```

here training accuracy is 85% and test accuracy is 81%, there is no more far, allmost similiar so we can say this is overfitting.

```python
input_data=(63,1,0,130,330,1,0,132,1,1.8,2,3,3)

# cahnge the input data to numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for only on instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==0):
  print("The person does not have a Heart Diseaes")
else:
  print("The person has Heart Diseases")
```

```python
import numpy as np  
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

```

Label Encoding:-

->Label Encoding refers to converting the labels into a numeric form.

->Here Loan_Status is in Y and N form, so we will convert Y into 1 and N into 0.

```python
# label encoding
loan_dataset.replace({'Loan_Status':{'N':0,'Y':1}},inplace=True)
```

```python
print(loan_dataset)
```

1 ---->Approved

0----->Rejected

```python
#Dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
# replacing the 3+ value into 4 or other
loan_dataset=loan_dataset.replace(to_replace='3+',value=4)
```

```python
loan_dataset['Dependents'].value_counts()
```

Data Visualization :

```python
# education & Loan Status
sns.countplot(x='Education',hue='Loan_Status',data=loan_dataset)
```

```python
# marital status & Loan Status
sns.countplot(x='Married',hue='Loan_Status',data=loan_dataset)
```

```python
# Gender & Loan Status
sns.countplot(x='Gender',hue='Loan_Status',data=loan_dataset)
```

```python
# convert categorical columns to numerical values
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
loan_dataset.head()
     
```

```python
# separating the data and label
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y = loan_dataset['Loan_Status']
     
```

```python
print(X)
print(Y)
```

Train test split

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
     
```

Training the model :

Support Vector Machine

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support Vector Macine model
classifier.fit(X_train,Y_train)
```

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuray = accuracy_score(X_train_prediction,Y_train)
     
```

```python
print('Accuracy on training data : ', training_data_accuray)
```

```python
# accuracy score on training data
X_test_prediction = classifier.predict(X_test)
test_data_accuray = accuracy_score(X_test_prediction,Y_test)
     
```

```python
print('Accuracy on test data : ', test_data_accuray)
```
